```python
len(POI_BUA)
```

```python
#naics codes of placekeys with no subamenity
POI_BUA[POI_BUA['sub_amenity'].isnull()]['naics_code'].unique()
```

```python
#Erase the rows with no naics_code
POI_BUA = POI_BUA[POI_BUA.naics_code.notnull()].reset_index().drop('index', axis = 1)
```

```python
# Count the number of placekeys per amenity 
POI_per_ame = POI_BUA.groupby(['sub_amenity',])['placekey'].count().reset_index().sort_values(by='placekey', ascending = False).rename(columns = {'placekey':'count_POIs'}).reset_index().drop('index', axis = 1)
POI_per_ame.head(10)
```

```python
#Create a column with the first 3 digits (because some placekeys have just 3 digits) of the naics_code for every placekey and amenity name
amenities_name['naics_3'] = amenities_name['naics_code'].astype(str).str[:3].astype(int)
POI_BUA['naics_3'] = POI_BUA['naics_code'].astype(str).str[:3].astype(int)
amenities_name = amenities_name.merge(POI_per_ame[['sub_amenity', 'count_POIs']], on='sub_amenity', how = 'left')
amenities_name.head(2)
```

```python
#Associate an amenity name to the unregistered naics codes 
for i in range(len(POI_BUA)):
  #Extract the sub_amenity name of the trip. If it is Nan, then we will have to look for an approximate naics code
  aux=POI_BUA.at[i, 'sub_amenity']
  if (isinstance(aux, str)==False):
  # Extracting the 3 digits naics code of the trip 
    naics=POI_BUA.at[i, 'naics_3']
  
  # Take the sub amenities that coincide with that naics code
    possible_sub_amenities = amenities_name[(amenities_name['naics_3'] == naics)]

  # Condition true if any sub_amenity meet the condition
    if (possible_sub_amenities.empty == False):
    # Weighted randomization association based on Safegraph’s registered POIs. The first argument is the list of objects from which it must select one and the second argument is 
    # the list of weights for each object of the list. Therefore, the sub_amenity with more POIs will be selected.
      POI_BUA.at[i, 'sub_amenity'] = random.choices(list(possible_sub_amenities['sub_amenity']), list(possible_sub_amenities['count_POIs']))[0]

POI_BUA.head(10)
```

```python
#Import amenities classification (sub-top-purpose) to add to each sub_amenity its top_amenity and travel_purpose
amenities_classification = pd.read_csv ('drive/MyDrive/Amenities/data/Amenities_classification.csv', index_col = False)

POI_BUA= POI_BUA.merge(amenities_classification[['travel_purpose', 'top_amenity', 'sub_amenity']], on='sub_amenity', how = 'left')
POI_BUA.head(10)
```

# SIZE INFO

```python
#Erase the rows with null area
trips_aux=POI_BUA[POI_BUA['wkt_area_sq_meters'].notnull()]
#Erase the POIs classified as SHARED POLYGON
trips_aux=POI_BUA[POI_BUA['polygon_class']=='OWNED_POLYGON']

#Get an array with the sub amenities name
sub_amenity_names=trips_aux['sub_amenity'].unique()

#Filter the outliers of the geometry file
trips_filt=pd.DataFrame()
for i in range(len(sub_amenity_names)):
  #Extract the sub amenity name from the array
  sub_amenity=sub_amenity_names[i]
  #Extract the trips to that type of sub amenity
  trips_subam=trips_aux[trips_aux['sub_amenity']==sub_amenity].reset_index()
  #Calculate the  scores
  trips_subam['z_scores'] =stats.zscore(trips_subam['wkt_area_sq_meters'])
  #Take out the outliers
  trips_temp = trips_subam[np.abs(trips_subam['z_scores']) < 3].reset_index().drop('index', axis = 1)
  #Make a dataframe with the filtered trips
  trips_filt=pd.concat([trips_filt, trips_temp])

trips_filt.head(2)
```

```python
len(trips_filt)
```

```python
sz_info_mean = trips_filt.groupby(['travel_purpose', 'top_amenity', 'sub_amenity'])['wkt_area_sq_meters'].mean().reset_index().rename(columns = {'wkt_area_sq_meters': 'mean_sqm'}).sort_values(by = 'mean_sqm', ascending = False)
sz_info_mean.head()
```

```python
sz_info_median = trips_filt.groupby(['travel_purpose', 'top_amenity', 'sub_amenity'])['wkt_area_sq_meters'].median().reset_index().rename(columns = {'wkt_area_sq_meters': 'median_sqm'}).sort_values(by = 'median_sqm', ascending = False)
sz_info_median.head()
```

```python
sz_info=sz_info_mean.merge(sz_info_median[['sub_amenity', 'median_sqm']], how= 'left', on= 'sub_amenity')
sz_info.head(2)
```

```python
sz_info['min_sqm']=""
sz_info['max_sqm']=""

for i in range(len(sz_info)):
  #Extract the sub_amenity name of the trip. If it is Nan, then we will have to look for an approximate naics code
  name=sz_info.at[i, 'sub_amenity']
  list_POIs=trips_filt[(trips_filt['sub_amenity']==name)]
  min=list_POIs['wkt_area_sq_meters'].min()
  max=list_POIs['wkt_area_sq_meters'].max()

  sz_info.at[i, 'min_sqm'] = min
  sz_info.at[i, 'max_sqm'] = max

sz_info.head(2)
```

```python
sz_info.columns
```

```python
#Total sqm by sub_amenity
total_sqm=trips_filt.groupby(['travel_purpose', 'top_amenity', 'sub_amenity']).agg({'wkt_area_sq_meters': 'sum'}).reset_index().rename(columns = {'wkt_area_sq_meters': 'total_sq_meters'})
total_sqm.head(2)
```

```python
sz_info=sz_info.merge(total_sqm[['sub_amenity', 'total_sq_meters']], how= 'left', on= 'sub_amenity')
sz_info
```
